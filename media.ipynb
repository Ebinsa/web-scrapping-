```python
import matplotlib.pyplot as plt
cf['Ad Volume'] = cf['Ad Volume'].astype(float)

# Identify potential outliers
q1 = cf['Ad Volume'].quantile(0.25)
q3 = cf['Ad Volume'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = cf[(cf['Ad Volume'] < lower_bound) | (cf['Ad Volume'] > upper_bound)]
print("Potential outliers:")
print(outliers)

# Visualize the data with outliers
plt.figure(figsize=(8, 6))
plt.boxplot(cf['Ad Volume'])
plt.title('Ad Volume Distribution')
plt.xlabel('Data')
plt.ylabel('Ad Volume')
plt.show()
```

```python
num_brands = cf['brand Name'].nunique()
print(f"Number of unique brands: {num_brands}")

# Top 3 brands by ad volume
top_brands = cf.nlargest(3, 'Ad Volume')
print("\nTop 3 brands by ad volume:")
print(top_brands[['brand Name', 'Ad Volume']])
```

```python
from selenium.webdriver.common.by import By
chrome_driver="/Users/ebinsam/Downloads/chromedriver "
download_file_path="/Users/ebinsam/Downloads"


options= webdriver.ChromeOptions()
options.add_argument("--disable-infobars")
options.add_argument("start-maximized")
options.add_argument("--disable-extensions" )
options.add_argument("--disable-nortifications")
options.add_argument('--ignore-certificate-errors')
options.add_experimental_option('prefs', {'download.default_directory': download_file_path, 'download.prompt_for_download': False , "download.directory_upgrade":True})


driver=webdriver.Chrome( options=options)

driver.get(url)
time.sleep(4)
```

```python
dropdown = driver.find_element(By.ID, "genres")
```

```python
options = dropdown.find_elements(By.TAG_NAME, "option")
```

```python
for option in options:
    if option.text == "KERALA":
        option.click()
        break
```

```python
table = driver.find_element(By.ID, "channel_data")
```

```python
rows = table.find_elements(By.TAG_NAME, "tr")
```

```python
table_data = []
for row in rows:
    cells = row.find_elements(By.TAG_NAME, "td")
    row_data = [cell.text for cell in cells]
    # Process the row data as needed
    print(row_data)
    table_data.append(row_data)
```

```python
column_names = ['Rank', 'KERALA Channels', 'Weekly AMA']

# Create a list of tuples from the table data
data_tuples = [(table_data[i][0], table_data[i][1], table_data[i][2]) 
               for i in range(0, len(table_data)) 
               if len(table_data[i]) >= 3]
```

```python
Ef = pd.DataFrame(data_tuples, columns=column_names)
```

```python
Ef
```

```python
# Convert 'Weekly AMA' column to numeric data type
Ef['Weekly AMA'] = pd.to_numeric(Ef['Weekly AMA'], errors='coerce')

# Replace any remaining non-numeric values with NaN
Ef['Weekly AMA'] = Ef['Weekly AMA'].fillna(0)

# Calculate total viewership
total_viewership = Ef['Weekly AMA'].sum()

# Calculate market share
Ef['Market Share'] = Ef['Weekly AMA'] / total_viewership

print("\nMarket Share:")
print(Ef[['KERALA Channels', 'Market Share']])
```

```python
plt.figure(figsize=(10, 6))
plt.bar(Ef['KERALA Channels'], Ef['Weekly AMA'])
plt.xlabel('KERALA Channels')
plt.ylabel('Weekly AMA')
plt.title('Channel Viewership')
plt.xticks(rotation=45)
plt.show()
```
